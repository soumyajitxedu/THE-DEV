```python
import torch
import torch.nn as nn 
import torch.optim as optim
x = torch.tensor([[1.], [2.], [3.], [4.]])
y = torch.tensor([[2.], [4.], [6.], [8.]])
class squarenet(nn.Module):
    def __init__(self):
        super(squarenet,self).__init__()
        self.hidden = nn.Linear(1, 10)
        self.output = nn.Linear(10, 1)
        self.relu = nn.ReLU()
    def forward(self, x):
        x = self.hidden(x)
        x = self.relu(x)
        x = self.output(x)
        return x
    
model = squarenet()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
for epoch in range(10000):
    yp = model(x)
    loss = criterion(yp, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}/100, Loss: {loss.item()}")

#test
test_val = torch.tensor([[10.]])
predicted = model(test_val)
print(f"Input: {test_val.item()}, Predicted Output: {predicted.item():.2f}")
```
