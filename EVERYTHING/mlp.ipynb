```python
import torch
import torch.nn as nn
class mlp(nn.Module):
    def __init__(self,d_model,d_ff):
        super().__init__()
        self.fc1 = nn.Linear(d_model, d_ff)
        self.fc2 = nn.Linear(d_ff, d_model)
        self.relu = nn.ReLU()
    def forward(self,x):
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.relu(x)
        return x
mlp_instance = mlp(d_model=128, d_ff=256)
print(f"mlp: {mlp_instance}")
```
