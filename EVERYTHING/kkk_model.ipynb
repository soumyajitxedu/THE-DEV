```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

X = np.array([
    [1,7],
    [2,5],
    [3,8],
    [5,9],
    [6,10],
    [7,12]
])
y = np.array([1,0,1,0,1,0])
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

weight = float(input("enter the weight of food "))
size = float(input("enter the size of the food "))
prediction = model.predict(np.array([[weight, size]]))[0]
if prediction == 0:
    print("it maybe game")
else:
    print("i don't know")
```
