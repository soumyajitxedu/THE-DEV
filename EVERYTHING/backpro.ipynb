# simple ann but backprop

```python
import torch
import torch.nn as nn

# Simple 2 layer network
class TwoLayer(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1   = nn.Linear(1, 2)
        self.l2   = nn.Linear(2, 1)
        self.relu = nn.ReLU()
        self.sigmodi = nn.Sigmoid()
    def forward(self,x):
        x = self.relu(self.l1(x))
        x = self.l2(x)
        return x

model = TwoLayer()
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(),lr = 0.01)
x = torch.tensor([[1.0]])
yt = torch.tensor([[3.0]])
yp = model(x)
loss = criterion(yp,yt)
print(f"loss : {loss.item():.2f}")
loss.backward()
print("\nBackprop gradients:")
for name, param in model.named_parameters():
    print(f"{name:20} grad={param.grad.flatten()}")

# Update
optimizer.step()
optimizer.zero_grad()
```
