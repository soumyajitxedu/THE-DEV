Basic model with loss and mse


```python
import torch
x = torch.tensor([1,2,3,4,5,6])
y = torch.tensor([2,4,5,6,7,9])
w = torch.tensor(0.0)
b = torch.tensor(0.0)

y_p = w * x + b
loss = ((y_p - y) ** 2).mean()
print("predicted value", y_p)
print("actural value " , y)
print("loss",loss)

```
