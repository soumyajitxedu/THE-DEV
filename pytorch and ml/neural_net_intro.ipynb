### my frist neural network 

```python
import torch 
import torch.nn as nn
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(5,1)
    
    def forward(self, x):
        return self.fc(x)
model = MyModel()
print(model)
x = torch.tensor([1.,2.,3.,4.,5.])
yt = torch.tensor([2.,4.,6.,8.,10.])
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(),lr = 0.2)
for epoch in range(100):
    yp = model(x)
    loss = criterion(yt,yp)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"epoch{epoch}, loss={loss.item()}")
        print('yp=',yp.detach().numpy())
test = torch.tensor([5.])

```

```python
import torch
import torch.nn as nn 
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(1,1)
    
    def forward(self,x):
        return self.fc(x)

model = MyModel()
print(model)
x = torch.tensor([[1.],[2.],[3.],[4.],[5.]])
yt = torch.tensor([[2.],[4.],[6.],[8.],[10.]    ])
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(),lr = 0.2)
for epoch in range(100):
    yp = model(x)
    loss = criterion(yt,yp)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"epoch{epoch}, loss={loss.item():.2f}")
        print('yp=',yp.detach().numpy())
```
